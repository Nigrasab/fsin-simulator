```python
from google.colab import files
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# АВТОМАТИЧЕСКАЯ ЗАГРУЗКА
uploaded = files.upload()

if uploaded:
    file_name = list(uploaded.keys())[0]
    data = pd.read_csv(io.BytesIO(uploaded[file_name]))
    print(f"Файл {file_name} успешно загружен и прочитан.")
    display(data.head())
else:
    print("Вы не выбрали файл!")
```

```python
# РАСЧЕТ БАЛАНСА И АНТИСИММЕТРИИ
signals = data.iloc[:, 1:]
data['Total_Balance'] = signals.sum(axis=1)
stability_index = np.mean(np.abs(data['Total_Balance']))

print(f"\nИндекс структурной устойчивости: {stability_index:.6f}")
print("Показатель < 0.1 подтверждает эффективность Золотого вурфа (1.309).")
```

```python
# ВИЗУАЛИЗАЦИЯ
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
plt.plot(data.iloc[:, 0], data['Total_Balance'], color='crimson')
plt.axhline(0, color='black', linestyle='--')
plt.title('Динамика самокомпенсации сил (Баланс)')

plt.subplot(1, 2, 2)
sns.heatmap(signals.corr(), cmap='RdBu_r', center=0)
plt.title('Матрица антисимметричных корреляций')
plt.show()
```
