# Синхронный Анализ ФСИН
Загрузите `fsin_sphiral.csv` для верификации резонанса.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files

uploaded = files.upload()
if uploaded:
    fn = list(uploaded.keys())[0]
    data = pd.read_csv(io.BytesIO(uploaded[fn]))
    nodes = [c for c in data.columns if 'Node' in c]
    balance = data[nodes].sum(axis=1)
    print(f'Индекс устойчивости: {np.mean(np.abs(balance)):.18f}')
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1); plt.plot(balance, color='black'); plt.axhline(0, color='red', ls='--')
    plt.subplot(1, 2, 2); sns.heatmap(data[nodes].corr(), cmap='RdBu_r', center=0)
    plt.show()
```
