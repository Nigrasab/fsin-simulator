# Анализ Сфирального Резонанса
Загрузите файл `fsin_sphiral.csv` при запуске ячейки.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files

uploaded = files.upload()
if uploaded:
    fn = list(uploaded.keys())[0]
    data = pd.read_csv(io.BytesIO(uploaded[fn]))
    nodes = [c for c in data.columns if 'Node' in c]
    data['Total_Balance'] = data[nodes].sum(axis=1)
    idx = np.mean(np.abs(data['Total_Balance']))

    print(f'Индекс устойчивости: {idx:.18f}')
    plt.figure(figsize=(10,4))
    plt.plot(data['Total_Balance'], color='black')
    plt.axhline(0, color='red', ls='--')
    plt.title('Баланс сил в Colab')
    plt.show()
```
