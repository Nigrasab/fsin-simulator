# Анализ Сфирального Резонанса ФСИН
### Инструкция:
1. Запустите ячейку ниже.
2. Выберите файл `fsin_sphiral.csv`, скачанный из приложения Streamlit.
3. Система рассчитает индекс устойчивости и построит термокарту антисимметрии.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files

# 1. Загрузка данных
print('Ожидание выбора файла...')
uploaded = files.upload()

if not uploaded:
    print('Файл не выбран.')
else:
    fn = list(uploaded.keys())[0]
    data = pd.read_csv(io.BytesIO(uploaded[fn]))
    
    # 2. Определение колонок узлов
    nodes = [c for c in data.columns if 'Node' in c]
    if len(nodes) == 0:
        print('Ошибка: Колонки Node_ не найдены!')
    else:
        # 3. Расчет баланса
        data['Total_Balance'] = data[nodes].sum(axis=1)
        idx = np.mean(np.abs(data['Total_Balance']))
        
        print(f'\n--- РЕЗУЛЬТАТЫ ---')
        print(f'Обработано узлов: {len(nodes)}')
        print(f'Индекс устойчивости: {idx:.18f}')
        
        # 4. Визуализация
        plt.figure(figsize=(15, 6))
        
        # График баланса
        plt.subplot(1, 2, 1)
        plt.plot(data['Total_Balance'], color='black', lw=2)
        plt.axhline(0, color='red', ls='--', alpha=0.5)
        plt.title('Суммарный вектор сил (Баланс)')
        plt.grid(True, alpha=0.2)
        
        # Термокарта
        plt.subplot(1, 2, 2)
        sns.heatmap(data[nodes].corr(), cmap='RdBu_r', center=0)
        plt.title('Матрица корреляции (Антисимметрия)')
        
        plt.tight_layout()
        plt.show()
```
